## Comparación entre las formas de la serie de Fourier

La serie de Fourier es una herramienta fundamental en el análisis de señales, ya que permite representar cualquier señal periódica como una suma de funciones sinusoidales. Existen tres formas principales para expresar esta serie: **forma trigonométrica**, **forma exponencial** y **forma compacta**. Aunque todas describen lo mismo, difieren en la forma matemática de escribir la señal y en la interpretación de los coeficientes.

### 1. Serie de Fourier Trigonométrica

Esta forma representa la señal como una suma infinita de senos y cosenos. La fórmula general es:

$ x(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n \omega_0 t) + b_n \sin(n \omega_0 t) \right) $

- $a_0$ es el valor promedio (componente DC) de la señal.
- $a_n$ y $b_n$ son coeficientes reales que indican la amplitud de cada componente armónica.
- $\omega_0 = \frac{2\pi}{T}$ es la frecuencia fundamental.

Esta forma es fácil de interpretar porque cada término corresponde a una frecuencia específica, con una amplitud y una fase definida.

### 2. Serie de Fourier Exponencial

En esta versión, la señal se expresa usando exponenciales complejas, lo cual facilita muchos análisis matemáticos:

$ x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t} $

- Los $C_n$ son coeficientes complejos que contienen toda la información de amplitud y fase.
- Esta forma es más compacta y general que la trigonométrica, y se usa mucho en teoría de sistemas y procesamiento de señales.

### 3. Forma Compacta

El término "forma compacta" suele referirse simplemente a la forma exponencial, ya que es una versión más condensada y algebraicamente manejable. Se basa en el uso de números complejos para representar la señal de forma más eficiente.



### Semejanzas

- Todas representan señales periódicas.
- Permiten descomponer una señal en sumas de funciones más simples (senos, cosenos o exponenciales).
- Los coeficientes de cada forma están relacionados y se pueden convertir de una forma a otra.

### Diferencias

| Característica          | Trigonométrica                            | Exponencial                                | Compacta                             |
|-------------------------|--------------------------------------------|---------------------------------------------|---------------------------------------|
| Funciones base          | Senos y cosenos                           | Exponenciales complejas                     | Equivalente a la exponencial          |
| Coeficientes            | Reales ($a_n$, $b_n$)                     | Complejos ($C_n$)                           | Complejos                             |
| Interpretación física   | Más intuitiva                             | Más técnica, útil para análisis matemático  | Similar a la exponencial              |
| Simetría                | Explícita                                 | Implícita ($C_{-n} = C_n^*$ para señales reales) | Idem exponencial                  |

## Transformada de Fourier, DTFT y DFT: semejanzas y diferencias

Después de entender las series de Fourier para señales periódicas, es importante conocer cómo se analiza una señal **no periódica** o **discreta**. Aquí es donde aparecen la **Transformada de Fourier (FT)**, la **Transformada de Fourier en tiempo discreto (DTFT)** y la **Transformada Discreta de Fourier (DFT)**. Aunque tienen nombres parecidos, cada una se usa en un contexto diferente.

### 1. Transformada de Fourier (FT)

Se aplica a **señales continuas en el tiempo** que pueden ser **no periódicas**. Convierte la señal del dominio del tiempo al dominio de la frecuencia continua.

$ X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt $

- $x(t)$ es la señal en el tiempo.
- $X(f)$ es la representación en frecuencia.
- El espectro resultante es **continuo**.

Se usa mucho en física, comunicaciones y procesamiento de señales analógicas.

### 2. Transformada de Fourier en Tiempo Discreto (DTFT)

Se aplica a señales **discretas en el tiempo pero infinitas**. También entrega un espectro **continuo**, pero periódico.

$ X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n} $

- $x[n]$ es la señal discreta.
- $X(e^{j\omega})$ es su representación en frecuencia, periódica en $\omega$ con periodo $2\pi$.

La DTFT es muy útil en teoría de procesamiento digital de señales.

### 3. Transformada Discreta de Fourier (DFT)

Se aplica a señales **discretas y de duración finita**. A diferencia de la DTFT, el resultado es también **discreto** (en frecuencia).

$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} kn} $

- $x[n]$ es la señal de entrada, con $N$ muestras.
- $X[k]$ representa las componentes de frecuencia discretas.
- Se implementa fácilmente en computador y es base para algoritmos como la FFT.

### Comparación general

| Aspecto                         | FT                           | DTFT                                | DFT                                |
|----------------------------------|-------------------------------|--------------------------------------|-------------------------------------|
| Tipo de señal                    | Continua, infinita           | Discreta, infinita                  | Discreta, finita                    |
| Resultado (espectro)            | Continuo                     | Continuo, pero periódico             | Discreto y periódico                |
| Aplicación                      | Señales analógicas           | Señales digitales teóricas          | Señales digitales reales (cómputo) |
| Implementación en computador    | No                           | No directamente                     | Sí                                  |

## Transformadas de Fourier – Paso a Paso

### a)   $ x(t) = e^{-a|t|}, \; a > 0 $

$$
X(f) = \int_{-\infty}^{\infty} e^{-a|t|} e^{-j 2\pi f t} dt
$$

Partimos en dos integrales:

$$
X(f) = \int_{-\infty}^{0} e^{(a - j2\pi f)t} dt + \int_{0}^{\infty} e^{-(a + j2\pi f)t} dt
$$

Evaluamos ambas:

$$
X(f) = \frac{1}{a - j2\pi f} + \frac{1}{a + j2\pi f}
= \frac{2a}{a^2 + (2\pi f)^2}
$$

---

### b) $ x(t) = \cos(w_c t) $

Usamos la forma exponencial:

$$
\cos(w_c t) = \frac{1}{2}(e^{j w_c t} + e^{-j w_c t})
$$

Entonces:

$$
X(f) = \frac{1}{2} \left[ \delta\left(f - \frac{w_c}{2\pi}\right) + \delta\left(f + \frac{w_c}{2\pi}\right) \right]
$$

---

### c) $ x(t) = \sin(w_s t) $

Forma exponencial:

$$
\sin(w_s t) = \frac{1}{2j}(e^{j w_s t} - e^{-j w_s t})
$$

Transformada:

$$
X(f) = \frac{1}{2j} \left[ \delta\left(f - \frac{w_s}{2\pi}\right) - \delta\left(f + \frac{w_s}{2\pi}\right) \right]
$$

---

### d) $x(t) = f(t)\cos(w_c t) $

No se puede resolver sin conocer la forma exacta de \( f(t) \).  
En general, este caso implica una convolución en frecuencia, pero como **no se deben usar propiedades**, no se puede continuar directamente.

---

### e) $ x(t) = e^{-a|t|^2}, \; a > 0 $

Transformada de una gaussiana:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j 2\pi f t} dt
$$

Resultado conocido:

$$
X(f) = \sqrt{\frac{\pi}{a}} \, e^{-\frac{(\pi f)^2}{a}}
$$

---

### f) $ x(t) = A \, \text{rect}_d(t) $

La función \( \text{rect}_d(t) \) está definida en el intervalo \( [-d/2, d/2] \):

$$
X(f) = \int_{-d/2}^{d/2} A \, e^{-j 2\pi f t} dt
$$

Evaluamos:

$$
X(f) = A \cdot \frac{\sin(\pi f d)}{\pi f} = A d \cdot \text{sinc}(f d)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

## **Solución de Problemas usando Transformada de Fourier**

A continuación resolveremos los problemas planteados aplicando propiedades de la Transformada de Fourier.


### **Problema a):** ℱ{e^{-jw₁t}cos(wₐt)}

**Paso 1:** Aplicamos identidad de Euler al coseno:
$$
\cos(w_c t) = \frac{e^{jw_c t} + e^{-jw_c t}}{2}
$$

**Paso 2:** Multiplicamos por el término exponencial:
$$
e^{-jw_1 t}\cos(w_c t) = \frac{e^{j(w_c - w_1)t} + e^{-j(w_c + w_1)t}}{2}
$$

**Paso 3:** Aplicamos la transformada de Fourier a cada término:
$$
\mathscr{F}\{e^{j(w_c - w_1)t}\} = 2\pi \delta(w - (w_c - w_1))
$$
$$
\mathscr{F}\{e^{-j(w_c + w_1)t}\} = 2\pi \delta(w + (w_c + w_1))
$$

**Paso 4:** Combinamos resultados:
$$
\boxed{\pi \left[\delta(w - (w_c - w_1)) + \delta(w + (w_c + w_1))\right]}
$$

------------------------------------------------------------------------

### **Problema b):** ℱ{u(t)cos²(wₐt)}

**Paso 1:** Aplicamos identidad trigonométrica:
$$
\cos^2(w_c t) = \frac{1 + \cos(2w_c t)}{2}
$$

**Paso 2:** Separamos en dos términos:
$$
u(t)\cos^2(w_c t) = \frac{u(t)}{2} + \frac{u(t)\cos(2w_c t)}{2}
$$

**Paso 3:** Transformada del escalón unitario:
$$
\mathscr{F}\{u(t)\} = \pi \delta(w) + \frac{1}{jw}
$$

**Paso 4:** Transformada del término modulado:
$$
\mathscr{F}\{u(t)\cos(2w_c t)\} = \frac{1}{2} \left[\pi \delta(w-2w_c) + \pi \delta(w+2w_c)\right] + \frac{1}{2j(w-2w_c)} + \frac{1}{2j(w+2w_c)}
$$

**Paso 5:** Combinamos todos los términos:
$$
\boxed{\frac{\pi}{2}\delta(w) + \frac{1}{2jw} + \frac{\pi}{4}\left[\delta(w-2w_c)+\delta(w+2w_c)\right] + \frac{1}{4j(w-2w_c)} + \frac{1}{4j(w+2w_c)}}
$$

--------------------------------------------------------------------

### **Problema c):** ℱ⁻¹{7/(w²+6w+45) * 10/(8+jw/3)²}

**Paso 1:** Completamos cuadrados en el primer denominador:
$$
w^2 + 6w + 45 = (w + 3)^2 + 36
$$

**Paso 2:** Identificamos la transformada inversa del primer término:
$$
\frac{7}{(w+3)^2 + 36} \Rightarrow \frac{7}{6}e^{-3t}\sin(6t)u(t)
$$

**Paso 3:** Segundo término (forma estándar):
$$
\frac{10}{(8 + jw/3)^2} \Rightarrow 30t e^{-8t/3}u(t)
$$

**Paso 4:** Aplicamos propiedad de convolución:
$$
\mathscr{F}^{-1}\{F(w)*G(w)\} = 2\pi f(t)g(t)
$$

**Resultado final:**
$$
\boxed{2\pi \left(\frac{7}{6}e^{-3t}\sin(6t)u(t)\right) \cdot \left(30t e^{-8t/3}u(t)\right)}
$$

-----------------------------------------------------------------------------


### **Problema d):** ℱ{3t³}

**Paso 1:** Propiedad de derivada en frecuencia:
$$
\mathscr{F}\{t^n f(t)\} = j^n \frac{d^n}{dw^n}F(w)
$$

**Paso 2:** Para f(t) = 1 (constante):
$$
\mathscr{F}\{1\} = 2\pi \delta(w)
$$

**Paso 3:** Aplicamos para n = 3:
$$
\mathscr{F}\{t^3\} = j^3 \frac{d^3}{dw^3}(2\pi \delta(w)) = -2\pi j \delta'''(w)
$$

**Paso 4:** Multiplicamos por 3:
$$
\boxed{-6\pi j \delta'''(w)}
$$

-----------------------------------------------------------------------------

### **Problema e):** Sumatoria periódica

**Paso 1:** Identificamos transformadas inversas:
1. $\frac{1}{a^2 + w^2} \Rightarrow \frac{\pi}{a}e^{-a|t|}$
2. $\frac{1}{a + jw} \Rightarrow e^{-at}u(t)$

**Paso 2:** Aplicamos propiedad de periodicidad:
$$
\sum_{n=-\infty}^{\infty} F(w - n\omega_0) \Rightarrow f(t) \sum_{k=-\infty}^{\infty} \delta(t - kT)
$$

**Paso 3:** Aplicamos a cada término de la sumatoria:
1. Para $\frac{1}{a^2+(w-nω_0)^2}$:
$$
\frac{B\pi}{aT}e^{-a|t|} \sum_{k=-\infty}^{\infty} \delta(t - kT)
$$

2. Para $\frac{1}{a+j(w-nω_0)}$:
$$
\frac{B}{T}e^{-at}u(t) \sum_{k=-\infty}^{\infty} \delta(t - kT)
$$

**Resultado final:**
$$
\boxed{\frac{B\pi}{aT}e^{-a|t|} \sum_{k=-\infty}^{\infty} \delta(t - kT) + \frac{B}{T}e^{-at}u(t) \sum_{k=-\infty}^{\infty} \delta(t - kT)}
$$








– Consulte en que consiste la modulación por amplitud por deteccion coherente y sus aplicaciones. Genere un ejemplo

La modulación por amplitud (AM) es una técnica de comunicación donde la amplitud de una portadora senoidal varía en función de una señal de mensaje.

*Detección coherente: es un método de
demodulación donde la portadora del receptor está sincronizada en frecuencia y fase con la del transmisor. Esto permite recuperar correctamente la señal original.

Aplicaciones principales:

*Radiodifusión AM.

*Comunicaciones de radio de larga distancia.

**Comunicaciones analógicas en aviones.

*Sistemas de radiofrecuencia en telefonía antigua.



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Fs = 1000  # Frecuencia de muestreo (Hz)
T = 1      # Tiempo total (s)
t = np.linspace(0, T, int(Fs * T), endpoint=False)

# Definición de señales de mensaje
def rectangular_pulse(t, width=0.2):
    return np.where((t % 1) < width, 1.0, 0.0)

def coseno_mensaje(t, f_m):
    return np.cos(2 * np.pi * f_m * t)

# Parámetros de modulación
f_c = 100  # Frecuencia de portadora (Hz)
m = 0.8    # Índice de modulación (ajustable por el usuario)
f_m = 5    # Frecuencia de mensaje (para el coseno)

# Señales de mensaje
mensaje_rect = rectangular_pulse(t)
mensaje_cos = coseno_mensaje(t, f_m)

# Señales moduladas
portadora = np.cos(2 * np.pi * f_c * t)
AM_rect = (1 + m * mensaje_rect) * portadora
AM_cos = (1 + m * mensaje_cos) * portadora

# FFT (frecuencia)
def plot_fft(signal, Fs, title):
    N = len(signal)
    freqs = np.fft.rfftfreq(N, d=1/Fs)
    fft_signal = np.abs(np.fft.rfft(signal)) / N
    plt.plot(freqs, fft_signal)
    plt.title(title)
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('Amplitud')
    plt.grid(True)

# Gráficas
plt.figure(figsize=(12, 8))

# Tiempo - Rectangular
plt.subplot(2, 2, 1)
plt.plot(t, AM_rect)
plt.title("AM con mensaje Rectangular (Tiempo)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True)

# Frecuencia - Rectangular
plt.subplot(2, 2, 2)
plot_fft(AM_rect, Fs, "AM con mensaje Rectangular (Frecuencia)")

# Tiempo - Coseno
plt.subplot(2, 2, 3)
plt.plot(t, AM_cos)
plt.title("AM con mensaje Coseno (Tiempo)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True)

# Frecuencia - Coseno
plt.subplot(2, 2, 4)
plot_fft(AM_cos, Fs, "AM con mensaje Coseno (Frecuencia)")

plt.tight_layout()
plt.show()
```

## **Ecuaciones de Modulación AM**

La modulación de amplitud (AM) se representa mediante:

$$
\begin{align*}
y(t) &= \left(1+\frac{m(t)}{A_c}\right) A_c\cos(2\pi F_ct) \\
&= A_c\cos(2\pi F_ct)+m(t)\cos(2\pi F_ct)
\end{align*}
$$

```python
# Paso 1: Instalar e importar librerías (ejecutar primero)
!pip install librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.signal import find_peaks
import librosa
from IPython.display import Audio

# Paso 2: Subir tu archivo WAV
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Configuración (puedes modificar estos valores)
start_time = 20  # Segundo inicial del audio
duration = 5     # Duración a analizar (segundos)
Fc = 10000       # Frecuencia portadora (Hz)
indice_mod = 0.7 # Índice de modulación (0 a 1)

# --------------------------------------------------
# NO NECESITAS MODIFICAR NADA DEBAJO DE ESTA LÍNEA
# --------------------------------------------------

# Cargar el archivo de audio
audio, sr = librosa.load(filename, sr=None, offset=start_time, duration=duration)
print(f"\nAudio cargado: {filename}")
print(f"Duración: {len(audio)/sr:.2f}s | Frecuencia de muestreo: {sr}Hz")

# Escuchar el segmento de audio
print("\nReproduciendo audio original...")
display(Audio(audio, rate=sr))

# Normalizar señal
m_t = audio / np.max(np.abs(audio))
t = np.linspace(0, duration, len(audio), endpoint=False)

# Visualización de la señal original
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, m_t, color='b')
plt.title('Señal de Audio Original (Mensaje)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

# Espectro de frecuencia original
N = len(audio)
xf = fftfreq(N, 1/sr)[:N//2]
yf = fft(m_t)
plt.subplot(3, 1, 2)
plt.plot(xf, 2/N * np.abs(yf[0:N//2]), color='g')
plt.title('Espectro de Frecuencia del Mensaje')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid()
plt.xlim(0, 5000)

# Generar señal portadora
c_t = np.cos(2 * np.pi * Fc * t)

# Señal modulada AM
y_t = (1 + indice_mod * m_t) * c_t

# Escuchar señal modulada (opcional)
print("\nReproduciendo señal modulada AM...")
display(Audio(y_t, rate=sr))

# Visualización de la señal modulada
plt.subplot(3, 1, 3)
plt.plot(t[:1000], y_t[:1000], color='r')  # Mostrar solo primeros milisegundos
plt.title('Señal Modulada AM (Zoom primeros 1000 puntos)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()

# Espectro de la señal modulada
plt.figure(figsize=(12, 4))
yf_am = fft(y_t)
plt.plot(xf, 2/N * np.abs(yf_am[0:N//2]), color='purple')
plt.title('Espectro de la Señal Modulada AM')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid()
plt.xlim(Fc-3000, Fc+3000)  # Zoom alrededor de la portadora
plt.show()

# Análisis de componentes frecuenciales
peaks, _ = find_peaks(2/N * np.abs(yf[0:N//2]), height=0.01, distance=50)
frecs_principales = xf[peaks]
amplitudes_principales = (2/N * np.abs(yf[0:N//2]))[peaks]

print("\nRESULTADOS DEL ANÁLISIS:")
print(f"Portadora: {Fc} Hz")
print("\nComponentes principales del mensaje:")
for f, a in zip(frecs_principales[:5], amplitudes_principales[:5]):
    print(f"- {f:.1f} Hz (amplitud: {a:.4f})")

print("\nBandas laterales generadas:")
for f in frecs_principales[:5]:
    print(f"- {Fc-f:.1f} Hz y {Fc+f:.1f} Hz")
```

```python
# %% [markdown]
"""
## **Simulación Completa de Modulación AM con Demodulación Coherente**
"""

# %%
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# %% [markdown]
"""
### **1. Configuración de Parámetros**
"""
# %%
# Parámetros fundamentales
Fc = 1000    # Frecuencia portadora (Hz)
Fm = 100     # Frecuencia del mensaje (Hz)
Fs = 10*Fc   # Frecuencia de muestreo (Hz)
duration = 5/Fm  # Duración en múltiplos del período del mensaje
Im = 0.8     # Índice de modulación (0 < Im ≤ 1)

# Vector de tiempo
t = np.arange(0, duration, 1/Fs)

# %% [markdown]
"""
### **2. Generación de Señales**
"""
# %%
# Señal mensaje (usando sinc para mejor visualización)
m_t = np.sinc(2*np.pi*Fm*(t-0.5*duration)/np.pi)

# Calcular amplitud de portadora según índice de modulación
Ac = np.max(np.abs(m_t))/Im

# Señal portadora
c_t = Ac * np.sin(2*np.pi*Fc*t)

# Señal modulada AM
y_t = (1 + m_t/Ac) * c_t

# %% [markdown]
"""
### **3. Visualización en Dominio Temporal**
"""
# %%
plt.figure(figsize=(12, 8))

# Señal mensaje
plt.subplot(3, 1, 1)
plt.plot(t, m_t, 'b')
plt.title('Señal Mensaje $m(t)$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

# Señal portadora
plt.subplot(3, 1, 2)
plt.plot(t, c_t, 'r')
plt.title('Señal Portadora $c(t)$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

# Señal modulada AM
plt.subplot(3, 1, 3)
plt.plot(t, y_t, 'g')
plt.title('Señal Modulada AM $y(t)$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

plt.tight_layout()
plt.show()

# %% [markdown]
"""
### **4. Análisis Espectral**
"""
# %%
# Cálculo de FFTs
N = len(t)
freqs = fftfreq(N, 1/Fs)

M_f = fft(m_t)
C_f = fft(c_t)
Y_f = fft(y_t)

# Visualización
plt.figure(figsize=(12, 8))

# Espectro del mensaje
plt.subplot(3, 1, 1)
plt.plot(freqs[:N//2], np.abs(M_f[:N//2]), 'b')
plt.title('Espectro de la Señal Mensaje $|M(f)|$')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.xlim(0, 2*Fc)
plt.grid()

# Espectro de la portadora
plt.subplot(3, 1, 2)
plt.plot(freqs[:N//2], np.abs(C_f[:N//2]), 'r')
plt.title('Espectro de la Portadora $|C(f)|$')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.xlim(0, 2*Fc)
plt.grid()

# Espectro de la señal AM
plt.subplot(3, 1, 3)
plt.plot(freqs[:N//2], np.abs(Y_f[:N//2]), 'g')
plt.title('Espectro de la Señal AM $|Y(f)|$')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.xlim(0, 2*Fc)
plt.grid()

plt.tight_layout()
plt.show()

# %% [markdown]
"""
### **5. Demodulación Coherente**
"""
# %%
# Etapa 1: Multiplicación por portadora local
d_t = y_t * c_t

# Etapa 2: Filtrado pasa bajas (en frecuencia)
D_f = fft(d_t)
f_cutoff = 2*Fm  # Frecuencia de corte
D_f[np.abs(freqs) > f_cutoff] = 0  # Filtrado ideal

# Etapa 3: Transformada inversa
m_r_t = np.real(np.fft.ifft(D_f)) * 2/Ac  # Recuperación con factor de escala

# %% [markdown]
"""
### **6. Visualización de la Demodulación**
"""
# %%
plt.figure(figsize=(12, 8))

# Señal después de multiplicación
plt.subplot(3, 1, 1)
plt.plot(t, d_t)
plt.title('Señal después de Multiplicación $d(t) = y(t) \cdot c(t)$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid()

# Espectro después de multiplicación
plt.subplot(3, 1, 2)
plt.plot(freqs[:N//2], np.abs(D_f[:N//2]))
plt.title('Espectro después de Filtrado')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.xlim(0, 2*Fc)
plt.grid()

# Señal demodulada recuperada
plt.subplot(3, 1, 3)
plt.plot(t, m_r_t, 'b', label='Demodulada')
plt.plot(t, m_t, 'r--', alpha=0.5, label='Original')
plt.title('Comparación: Señal Original vs Demodulada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```

Distorsión Total de Armónicos (THD)

El THD mide la distorsión armónica presente en una señal y se define como:


## Fórmula de THD

$$
THD = \frac{\sqrt{\sum_{h=2}^\infty V_h^2}}{V_1} \times 100\%
$$

donde $V_1$ es la amplitud de la fundamental y $V_h$ son las amplitudes de los armónicos.

## Factor de Potencia (FP)
El factor de potencia relaciona la potencia real con la potencia aparente:


$$
FP = \frac{P}{S} = \frac{P}{V_{rms}I_{rms}}
$$

### En presencia de armónicos:
$$
FP = \frac{1}{\sqrt{1 + THD^2}} \times \cos(\phi_1)
$$

donde $\phi_1$ es el ángulo de fase de la fundamental.


```python
#implementacion en python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import trapezoid # Changed from simps

# Parámetros de simulación
fs = 10000       # Frecuencia de muestreo (Hz)
duration = 0.1   # Duración de simulación (s)
f = 50           # Frecuencia fundamental (Hz)
t = np.linspace(0, duration, int(fs*duration), endpoint=False)

def calculate_thd(signal, fs, f):
    """Calcula THD a partir de la FFT de una señal"""
    N = len(signal)
    yf = fft(signal)[:N//2]
    xf = fftfreq(N, 1/fs)[:N//2]

    # Encontrar fundamental y armónicos
    fundamental_idx = np.argmax(np.abs(yf[:int(1.5*f*N/fs)]))  # Search range based on fundamental freq
    V1 = np.abs(yf[fundamental_idx])

    # Identify harmonic indices (up to a reasonable limit, e.g., 20th harmonic or 2kHz)
    harmonic_indices = [i for i in range(len(xf))
                       if (i != fundamental_idx) and (xf[i] > f*1.5) and (xf[i] % f < f/4) and (xf[i] < 2000)] # Added tolerance and lower bound
    Vh = np.sqrt(np.sum(np.abs(yf[harmonic_indices])**2))

    return (Vh/V1)*100

def rectifier_output(R, C=0, plot=False):
    """Simula rectificador de onda completa con carga R o RC"""
    # Generar señal de entrada
    Vin = 120*np.sqrt(2)*np.sin(2*np.pi*f*t)

    # Simular rectificación
    Vrect = np.abs(Vin)

    # Simular carga
    if C > 0:  # Carga RC
        tau = R*C
        Vout = np.zeros_like(Vrect)
        for i in range(1, len(Vout)):
            Vout[i] = Vout[i-1] + (Vrect[i] - Vout[i-1])*(1 - np.exp(-1/(fs*tau)))
    else:      # Carga resistiva
        Vout = Vrect

    # Calcular corriente
    Iout = Vout/R

    if plot:
        plt.figure(figsize=(12,4))
        plt.plot(t[:500], Vout[:500], label='Tensión salida')
        plt.plot(t[:500], Iout[:500], label='Corriente salida')
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Amplitud')
        plt.legend()
        plt.grid()
        plt.show()

    return Vout, Iout

# Ejemplo de cálculo para carga resistiva
Vout_r, Iout_r = rectifier_output(R=100, plot=True)
thd_v_r = calculate_thd(Vout_r, fs, f)
thd_i_r = calculate_thd(Iout_r, fs, f)
print(f"THD tensión (carga R): {thd_v_r:.2f}%")
print(f"THD corriente (carga R): {thd_i_r:.2f}%")

# Ejemplo para carga RC
Vout_rc, Iout_rc = rectifier_output(R=100, C=100e-6, plot=True)
thd_v_rc = calculate_thd(Vout_rc, fs, f)
thd_i_rc = calculate_thd(Iout_rc, fs, f)
print(f"THD tensión (carga RC): {thd_v_rc:.2f}%")
print(f"THD corriente (carga RC): {thd_i_rc:.2f}%")
```

3. Análisis de Resultados
Caso 1: Carga Resistiva (R = 100Ω)
-THD típico: 45-50%

-Formas de onda:

Tensión y corriente perfectamente rectificadas

Alto contenido armónico

-Factor de potencia: Cercano a 1 (poca distorsión de fase)

Caso 2: Carga RC (R = 100Ω, C = 100μF)
-THD típico: 70-120%

*Efectos observados:

El capacitor reduce el rizado pero aumenta los picos de corriente

Mayor distorsión armónica en corriente

Factor de potencia disminuye por:

*Mayor THD

*Desfase introducido por el capacitor

```python
#simulación Completa con Variación de Parámetros
# Barrido de valores R y C
R_values = [50, 100, 200]
C_values = [0, 50e-6, 100e-6, 200e-6]

results = []

for R in R_values:
    for C in C_values:
        Vout, Iout = rectifier_output(R, C)
        thd_v = calculate_thd(Vout, fs, f)
        thd_i = calculate_thd(Iout, fs, f)

        # Cálculo factor de potencia aproximado
        FP = 1/np.sqrt(1 + (thd_i/100)**2)  # Asumiendo cos(φ)=1

        results.append({
            'R': R,
            'C': C,
            'THD_V': thd_v,
            'THD_I': thd_i,
            'FP': FP
        })

# Mostrar resultados
import pandas as pd
df = pd.DataFrame(results)
print(df.pivot(index='R', columns='C', values=['THD_V', 'THD_I', 'FP']))
```

```python
#Visualización de Resultados
# Gráfico de THD vs parámetros
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
for R in R_values:
    subset = df[df['R'] == R]
    plt.plot(subset['C']*1e6, subset['THD_V'], 'o-', label=f'R={R}Ω')
plt.xlabel('Capacitancia (μF)')
plt.ylabel('THD Tensión (%)')
plt.legend()
plt.grid()

plt.subplot(1,2,2)
for C in C_values[1:]:  # Excluir C=0
    subset = df[df['C'] == C]
    plt.plot(subset['R'], subset['FP'], 's--', label=f'C={C*1e6}μF')
plt.xlabel('Resistencia (Ω)')
plt.ylabel('Factor de Potencia')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

### 1. Efecto de la carga  
- **Cargas resistivas**:  
  - Producen menor THD (Distorsión Armónica Total)  
  - Generan mayor rizado en la señal de salida  

- **Cargas RC**:  
  - Reducen el rizado  
  - Aumentan el THD  
  - Degradan el Factor de Potencia (FP)

**Para mejorar el FP**:  
- Evitar valores excesivos de capacitancia  
- Implementar circuitos de corrección activa del factor de potencia
